```python
import pandas as pd
```

```python
month1 = pd.read_csv("cyclistic_042025.csv")
month2 = pd.read_csv("cyclistic_052025.csv")
month3 = pd.read_csv("cyclistic_062025.csv")
month4 = pd.read_csv("cyclistic_072025.csv")
month5 = pd.read_csv("cyclistic_082025.csv")
month6 = pd.read_csv("cyclistic_092025.csv")
month7 = pd.read_csv("cyclistic_102025.csv")
month8 = pd.read_csv("cyclistic_112025.csv")
month9 = pd.read_csv("cyclistic_122025.csv")
month10 = pd.read_csv("cyclistic_012026.csv")
month11 = pd.read_csv("cyclistic_022026.csv")
month12 = pd.read_csv("cyclistic_032026.csv")
```

```python
month1.info()
```

```python
all_trips = pd.concat([month1, month2, month3, month4, month5, month6, month7, month8, month9, month10, month11, month12], ignore_index=True)
```

```python
all_trips = all_trips.drop(columns=['start_lat', 'start_lng', 'end_lat', 'end_lng'], errors='ignore')
```

```python
all_trips['ride_length'] = pd.to_timedelta(all_trips['ride_length'])

# 2. Şimdi filtrenizi tekrar uygulayabilirsiniz.
# Dikkat: 0 sayısını da pd.Timedelta(0) olarak yazıyoruz ki elma ile elmayı kıyaslayabilelim.
all_trips_v2 = all_trips[
    (all_trips['start_station_name'] != "HQ QR") &
    (all_trips['ride_length'] >= pd.Timedelta(0))
].copy()
```

```python
all_trips_v2.groupby('member_casual')['ride_length'].agg(['mean', 'median', 'max', 'min'])
```

```python
all_trips_v2.groupby(['member_casual', 'day_of_week'])['ride_length'].mean()
```

```python
days_order = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
all_trips_v2['day_of_week'] = all_trips_v2['day_of_week'].astype(str)
all_trips_v2['day_of_week'] = all_trips_v2['day_of_week'].replace({
    1: "Sunday",
    2: "Monday",
    3: "Tuesday",
    4: "Wednesday",
    5: "Thursday",
    6: "Friday",
    7: "Saturday",
})
```

```python
all_trips_v2.groupby(['member_casual', 'day_of_week'])['ride_length'].mean()
```

```python
summary_stats = all_trips_v2.groupby(['member_casual', 'day_of_week']).agg(
    number_of_rides=('ride_id', 'count'),
    average_duration=('ride_length', 'mean')
).reset_index()
```

```python
counts = all_trips_v2.groupby(['member_casual', 'day_of_week'])['ride_length'].mean().reset_index()
counts.to_csv('avg_ride_length.csv', index=False)
```
